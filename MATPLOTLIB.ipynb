# PLOTING HISTOGRAMS

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_students_age=np.random.randint(18,45,(100))
py_students_age=np.random.randint(15,40,(100))
```

```python
print(ml_students_age)
print(py_students_age)
```

```python
plt.hist(ml_students_age)
plt.title("ML students age histogram")
plt.xlabel("students age category")
plt.ylabel("no. student age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8)
plt.title("ML students age histogram")
plt.xlabel("students age category")
plt.ylabel("no. student age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar",orientation="horizontal")
plt.title("ML students age histogram")
plt.xlabel("students age category")
plt.ylabel("no. student age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar",orientation="horizontal",color="r")
plt.title("ML students age histogram")
plt.xlabel("students age category")
plt.ylabel("no. student age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar",orientation="horizontal",color="r",label="ML Student")
plt.title("ML students age histogram")
plt.xlabel("students age category")
plt.ylabel("no. student age")
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,histtype="bar",orientation="vertical",color=["r","y"],label=["ML Student","PY Student"])
plt.title("ML students age histogram")
plt.xlabel("students age category")
plt.ylabel("no. student age")
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.figure(figsize=(16,9))
plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,histtype="bar",orientation="vertical",color=["r","y"],label=["ML Student","PY Student"])
plt.title("ML students age histogram")
plt.xlabel("students age category")
plt.ylabel("no. student age")
plt.legend()
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
bins=[15,20,25,30,35,40,45]
plt.figure(figsize=(16,9))
plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,histtype="bar",orientation="vertical",color=["r","y"],label=["ML Student","PY Student"])
plt.title("ML students age histogram")
plt.xlabel("students age category")
plt.ylabel("no. student age")
plt.legend()
plt.show()
```

# PIE CHART

```python
import matplotlib.pyplot as plt
```

```python
classes=["python","r","machine learning","artificial intelligence","data sciece"]
class_st=[45,15,35,25,30]
plt.pie(class_st,labels=classes)
plt.show()
```

```python
classes=["python","r","machine learning","artificial intelligence","data sciece"]
class_st=[45,15,35,25,30]
explode=[0.03,0,0.1,0,0]
colors=["c","b","r","y","g"]
textprops={"fontsize":15}
plt.pie(class_st,labels=classes,explode=explode,colors=colors,autopct="%0.1f%%",shadow=True,radius=1.4,startangle=90,textprops=textprops)

plt.show()
```

```python
plt.figure(figsize=(6,2))
wedgeprops={"linewidth":4,"width":2,"edgecolor":"k"}
plt.pie(
            class_st,
            labels=classes,
            explode=explode,
            colors=colors,
            autopct="%0.2f%%",
            pctdistance=0.6,
            shadow=True,
            labeldistance=1.4,
            startangle=270,
            radius=1.1,
            counterclock=True,
            wedgeprops=wedgeprops,
            textprops=textprops,
            center=(2,3),
            frame=True,
            rotatelabels=True)
plt.show()
```
